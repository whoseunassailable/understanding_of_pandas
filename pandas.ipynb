#### Creating a new column based on the data of other columns

#### Filtering the Data based on one condition

```python
# Using loc to see a particular value
legendary_pokemon = pookie.loc[(pookie["Legendary"] == True)]
legendary_pokemon
```

#### Filtering the Data based on multiple conditions

```python
# Please use circular brackets and mention a condition inside it like (pookie["Legendary"] == True)
legendary_pokemon_with_highest_total = legendary_pokemon.loc[(legendary_pokemon["Legendary"] == True) & (legendary_pokemon['total'] > 650)]
legendary_pokemon_with_highest_total
```

#### Using sort on top of Filtering

```python
sorted_legendary_pokemon = legendary_pokemon_with_highest_total.sort_values('total', ascending= True)
sorted_legendary_pokemon
```

```python
count_of_sorted_legendary_pokemon = sorted_legendary_pokemon.count().iloc[0]
count_of_sorted_legendary_pokemon
```

#### Let's save the legendary pokemon with highest total over 650
#### But while we do it, we need to make sure the indexing of the 
#### new data should not be equal to the old one

```python
# reset_index(drop = True) will reset the index to 0 and remove the old index column
# Setting inplace = True will help you to not to set a new variable to new df and then save it
legendary_pokemon_with_highest_total.reset_index(drop=True, inplace=True)
legendary_pokemon_with_highest_total.to_csv("legendary_pokemon_with_highest_values.csv")
```

# Filtering Data with the contains()

```python
pookie.loc[pookie['Name'].str.contains("Mega")] # Will display names with Mega in it
```

#### Using regex with contains()

```python
# You can do use the regular expression with the contains()
pookie.loc[pookie['Type 1'].str.contains("Fire|Grass",regex=True)] 
```

#### Setting the flag re.I = True to remove case sensitivity

```python
import re # importing the required re library

# If you search for fire|grass you will get null because it is case sensitive
pookie.loc[pookie['Type 1'].str.contains("fire|grass",flags=re.I, regex=True)] 
```

#### Using complex regex expression

```python
# Names that start with go
pookie.loc[pookie['Name'].str.contains("^go[a-z]*",flags=re.I, regex=True)] 
```

#### Conditional changes

```python
# So we are setting the Type 1 which are Fire as Flamer
pookie.loc[pookie['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
pookie
```

```python
# We are setting Fire Type as Legendary
pookie.loc[pookie['Type 1'] == 'Flamer', 'Legendary'] = True
pookie.loc[pookie['Type 1'] == 'Flamer'].head() # Displaying whether they are set to True or not
```

#### Setting multiple columns with single certain value conditional changes
#### Setting Generation and Name to Next Level

```python
pookie.loc[pookie['Legendary'] == True, ['Generation', 'Name']] = 'Next Level'
pookie
```

#### Setting multiple columns with multiple conditional values

```python
# Setting Generation = 6 and Name = Ultra
pookie.loc[pookie['total'] > 650, ['Generation', 'Name']] = [6, 'Ultra'] 
pookie.loc[pookie['total'] > 650]
```

#### Aggregate Statistics

```python
pookie = pd.read_csv('pokemon_data.csv') # Switching back to old data checkpoint
numeric_cols = pookie.select_dtypes(include=['number']).columns  # Select numeric columns
```

#### Getting mean of Type 1 column

```python
 pookie.groupby(['Type 1'])[numeric_cols].mean()
```

#### Getting sum of Type 1 column

```python
pookie.groupby(['Type 1'])[numeric_cols].sum()
```

#### Getting count of Type 1 column

```python
# pookie.groupby(['Type 1'])[numeric_cols].count()
pookie['count'] = 1 # adding a column name column and then giving all of them the value of 1 
pookie.groupby(['Type 1', 'Type 2'])[numeric_cols].count()
```

#### sort_values() on top of aggregated group by

```python
pookie.groupby(['Type 1'])[numeric_cols].mean().sort_values(['Defense', 'Attack'],ascending=[False, True])
```

#### Reading chunks of data

```python
for pookie in pd.read_csv('pokemon_data.csv', chunksize=5):
    print(pookie)
```

#### adding data to new dataframe

```python
new_pookie = pd.DataFrame(columns=pookie.columns)
pookie
for pookie in pd.read_csv('pokemon_data.csv', chunksize=5):
    results = pookie.groupby('Type 1')[numeric_cols].count()
    new_pookie = pd.concat([new_pookie, results])
new_pookie
```
